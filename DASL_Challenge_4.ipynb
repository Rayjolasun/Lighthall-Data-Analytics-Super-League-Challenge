## Task 2

**Clean the year column to get the appropriate year**

```python
# Extract the year using regular expressions
df['YEAR'] = df['YEAR'].str.extract(r'(\d{4})')
```

We extracted the year from the **YEAR column** of the DataFrame. The str.extract() function takes a regular expression as its argument and returns a new DataFrame with one column for each capturing group in the regular expression. In this case, the regular expression r'(\d{4})' matches a string of four digits. The expand=True argument to str.extract() tells pandas to return a new DataFrame with one column for each capturing group, even if there is only one capturing group. We were then able to create the appropriate YEAR column.

```python
# confirm that the appropriate year has been extracted

df.sample(10)
```

We randomly sample 10 rows from the DataFrame to confirm if the appropriate YEAR have been effected and the outcome could be seen above.

Then we converted the YEAR column of the DataFrame to a numeric type. The to_numeric() function takes the Year column as its argument and returns a new Year column with the converted values. The errors='coerce' argument tells pandas to convert all non-numeric values to NaN.

We then printed a concise summary of the new DataFrame which gives index dtype and columns, non-null values and memory usage.

## Task 5

**Get the director name and put it in a different column and also put the stars names in a separate column**

```python

# Extract the director names
df['Directors'] = df['STARS'].str.extract('Director[s]?:(.+?)(?=\s*\||Stars:|$)', flags=re.S, expand=False)
df['Directors'] = df['Directors'].str.replace('\n', '')

# Extract the stars names
stars1 = df['STARS'].str.extract('Star[s]?:(.+)', flags=re.S, expand=False)
stars2 = df['STARS'].str.extract('Stars:(.+)', flags=re.S, expand=False)

# Concatenate star names into a single column
df['Stars'] = stars1.fillna('').values.flatten() + stars2.fillna('').values.flatten()
df['Stars'] = df['Stars'].str.replace('\n', '')

# Remove leading/trailing whitespace from the extracted columns
df['Directors'] = df['Directors'].str.strip()
df['Stars'] = df['Stars'].str.strip()

# Fill NaN or null values or empty values with ''
df['Directors'] = df['Directors'].fillna("")
df['Stars'] = df['Stars'].fillna("")

# Insert the 'Directors' and 'stars' columns in the original STARS position
original_position = df.columns.get_loc('STARS')
df.insert(original_position, 'Directors', df.pop('Directors'))
df.insert(original_position + 1, 'Stars', df.pop('Stars'))

# Drop the 'STARS' column
df = df.drop('STARS', axis=1)

df.sample(20)
```

The code first extracted the directors from the STARS column using regular expressions. It then concatenate the star names into a single column, remove leading / trailing whitespace from the extracted columns, and fill NaN or null values or empty values with an empty string. Finally, it inserted the Directors and Stars columns in the original STARS position and drop the STARS column.

## Task 8

**Clean the gross column**

```python
has_nan = df['Gross'].isna().sum()
print("NaN values in 'Gross' column:", has_nan)
```

We counted the number of missing values in the Gross column of the DataFrame. The "isna()" function was used to identify these missing values. The "sum()" function was then used to sum the boolean Series returned by "isna()". We then printed the outcome which gives 6365 NaN values.

```python
df['Gross'] = df['Gross'].str.replace(',', '')  # Remove commas
df['Gross'] = df['Gross'].str.replace('[\$,M]', '', regex=True).astype(float) * 1e6

median_gross = df['Gross'].median()
df['Gross'].fillna(median_gross, inplace=True)  # Fill NaN with median
df['Gross'] = df['Gross'].astype(int)  # Convert to whole numbers (integer data type)


df.head()
```

In cleaning the Gross column, we removed commas, dollar signs and millions from the Gross column of the DataFrame.

We converted the values in the Gross column to floats and multiply them by 1 million.

We find the median of the Gross column and fill any missing values with the median.

We then converted the values in the Gross column to integers and printed the outcome which could be seen above.
